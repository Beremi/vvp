# 1. GIL v CPythonu

`GIL` (Global Interpreter Lock) je zámek v interpretu CPython, který zajišťuje, že Python byte-kód v jednom procesu vykonává vždy jen jedno vlákno najednou.

## 1.1 Proč GIL existuje?

Historicky zjednodušuje správu paměti a práci se sdílenými objekty (hlavně referenční počítání) bez jemného zamykání na úrovni všech objektů.

## 1.2 Jak GIL ovlivňuje vícevláknový kód?

- U čistě Pythonovského CPU-bound kódu typicky nezrychlíme výpočet přidáním vláken.
- Vlákna ale dávají smysl u I/O operací (disk, síť), kde se většinu času čeká.
- Kód v knihovnách napsaných v C/C++/Fortranu může běžet mimo GIL a využít více jader.

# 2. Jak psát nestriktně sekvenční programy v Pythonu

Prakticky používáme tři cesty:

- **Více procesů**: každý proces má vlastní interpreter a vlastní GIL.
  - `multiprocessing` pro lokální paralelizaci na jednom stroji.
  - `mpi4py` pro více procesů s explicitní komunikací (typicky i přes více uzlů).
- **Výpočty mimo Python byte-kód**: těžká práce probíhá v nativním kódu.
  - `numpy` (BLAS/LAPACK), `numba`, `cython`, část `scipy` a dalších knihoven.
  - V těchto částech lze využít více vláken a více jader.
- **Vlákna pro překrytí čekání (I/O-bound)**:
  - `threading` je vhodný hlavně pro čekání na síť, disk nebo jiné externí zdroje.

# Pojmenovvn promnnch, funkc, td, modul a balk

V Pythonu plat nsledujc konvence pojmenovn:

1. **Promnn**:
   - Pouvejte mal psmena a podtrtka pro oddlen slov(`snake_case`).
   - Napklad: jmeno_uzivatele, pocet_aut
2. **Funkce**:
   - Stejn jako promnn, pouijte mal psmena a podtrtka pro oddlen slov(`snake_case`).
   - Napklad: vypocet_obvodu, ziskej_data
3. **Tdy**:
   - Pouijte `PascalCase` (obas tak nazvn jako `CamelCase`), kde kad slovo zan velkm psmenem a nejsou zde dn podtrtka.
   - Napklad: MujSeznam, Osoba
4. **Moduly**:
   - Pouijte mal psmena a podtrtka pro oddlen slov, pokud je to nezbytn nutn.
   - Napklad: muj_modul, nastroje
5. **Balky**:
   - Pouijte mal psmena bez podtrtek.
   - Napklad: mujbalicek, databaze
6. **Konstanty**:
   - Pouijte velk psmena a podtrtka pro oddlen slov.
   - Napklad: MAX_POCET, RYCHLOST_SVETLA


**Obecn pravidla**:
   - Nzvy by mly bt popisn a ne pli dlouh - je teba najt rovnovhu mezi pehlednost kdu a jeho itelnost.
   - Nepouvejte nzvy, kter by mohly bt zamnny s klovmi slovy Pythonu(nap. `class `, `if `, `else ` atd.).
   - Pokud je dann problm typyck nzev, kter se kryje s Pythonovm klovm slovem, pouijte podtrtko na konci nzvu (nap. `class_`).
   - Bute jednotn v jazyce, kterm pojmenovvte promnn, funkce, tdy, moduly a balky. 
   - Pokud implementujete njak algoritmus, kter je popsn v njak literatue, pouijte nzvy promnnch, kter jsou pouity v tto literatue a v komench vysvtlete, co dan promnn reprezentuje (nap. `x` a `y` pro souadnice bodu). Je tak vhodn pidat odkaz na literaturu, kter popisuje dan algoritmus.
   - Nzvy, kter jsou viditeln pro uivatele jako veejn sti API, by mly dodrovat konvence, kter odrej pouit spe ne implementaci. - tedy mohou poruovat nkter pravidla, kter jsou uvedena ve.
   - Nzvy promnnch a funkc, kter nejsou ureny pro pouit (nap. jako soust API), zanaj podtrtkem. Napklad: `_nazev_promenne`, `_nazev_funkce`. Tato konvence je pomrn dleit, nebo Python neumouje "schovat" st implementace ped uivatelem.


# Docstringy

## Plain-text
Jednoduch a pmoar formt, kter se skld z prostho textu bez zvltnho formtovn. Jde snadno napsat, ale nemus poskytovat dostatenou strukturu pro velk projekty nebo nstroje, kter automaticky generuj dokumentaci.

**Pro ns preferovan varianta spolen s type hintingem.**

```python
def add(a: int | float, b : int | float) -> int | float:
    """
    Tato funkce bere jako vstup dv sla a vrt jejich souet.
    """
    return a + b
```

## reStructuredText (reST): 
Lehk znakovac jazyk, kter se bn pouv pro dokumentan etzce v Pythonu, zejmna pro velk projekty nebo pi pouit Sphinxu pro generovn dokumentace. M jednoduchou syntaxi a umouje bohat formtovn.

```python
def add(a, b):
    """
    Tato funkce bere jako vstup dv sla a vrt jejich souet.

    :param a: Prvn slo k setn
    :type a: int nebo float
    :param b: Druh slo k setn
    :type b: int nebo float
    :return: Souet sel a a b
    :rtype: int nebo float
    """
    return a + b
```

## Google styl: 
Formt, kter je jasn, strun a snadno iteln. Je populrn v komunit Pythonu a pouv ho mnoho projekt, vetn tch uvnit Google.

```python
def add(a, b):
    """
    Tato funkce bere jako vstup dv sla a vrt jejich souet.

    Args:
        a (int nebo float): Prvn slo k setn
        b (int nebo float): Druh slo k setn

    Returns:
        int nebo float: Souet sel a a b
    """
    return a + b
```

## NumPy styl: 
Formt, kter je speciln navren pro dokumentaci vdeckch projekt v Pythonu, jako jsou ty, kter pouvaj knihovny NumPy, SciPy nebo souvisejc. Je podobn Google stylu, ale m nkter dal konvence pro dokumentaci vdeckch funkc.

```python
def add(a, b):
    """
    Tato funkce bere jako vstup dv sla a vrt jejich souet

    Parametry
    ----------
    a : int nebo float
        Prvn slo k setn
    b : int nebo float
        Druh slo k setn

    Nvratov hodnoty
    -------
    int nebo float
        Souet sel a a b
    """
    return a + b
```

## DRY = "Don't Repeat Yourself" ("Neopakuj se") 

- Je to zsada, kter k, e kad sek informac nebo kdu by ml bt definovn pouze na jednom mst a neml by bt duplikovn. 
- DRY pomh vytvet krat, itelnj a udritelnj (se sna drbou) kd.


```python
# patn
# Data subjektu = [vha_kg, vka_m]
subjekt1 = [80, 1.62]
subjekt2 = [69, 1.53]
subjekt3 = [80, 1.66]
subjekt4 = [80, 1.79]
subjekt5 = [72, 1.60]

bmi_subjekt1 = int(subjekt1[0] / subjekt1[1]**2)
print("bmi {} = {}".format('subjekt1', bmi_subjekt1))

bmi_subjekt2 = int(subjekt2[0] / subjekt2[1]**2)
print("bmi {} = {}".format('subjekt2', bmi_subjekt2))

bmi_subjekt3 = int(subjekt3[0] / subjekt3[1]**2)
print("bmi {} = {}".format('subjekt3', bmi_subjekt3))

bmi_subjekt4 = int(subjekt4[0] / subjekt4[1]**2)
print("bmi {} = {}".format('subjekt4', bmi_subjekt4))

bmi_subjekt5 = int(subjekt5[0] / subjekt5[1]**2)
print("bmi {} = {}".format('subjekt5', bmi_subjekt5))
```

```python
# dobe
def vypocti_bmi(vaha_kg, vyska_m):
    return int(vaha_kg / vyska_m**2)


subjekty = {
    'subjekt1': [80, 1.62],
    'subjekt2': [69, 1.53],
    'subjekt3': [80, 1.66],
    'subjekt4': [80, 1.79],
    'subjekt5': [72, 1.60]
}

for jmeno, (vaha, vyska) in subjekty.items():
    bmi = vypocti_bmi(vaha, vyska)
    print(f"bmi {jmeno} = {bmi}")
```

## KISS (Keep It Simple, Stupid) 

je zsada pouvan v programovn a nvrhu systm, kter zdrazuje dleitost jednoduchosti a snadnho pochopen. KISS princip k, e vtina systm funguje lpe, pokud jsou jednoduch, ne kdy jsou sloit. Tm se sniuje riziko chyb, zjednoduuje drbu, zlepuje testovatelnost a itelnost kdu. 


**Udrovn jednoduchosti v kdu zahrnuje**:

1. Pouvn jasnch a srozumitelnch nzv promnnch a funkc.
2. Rozdlen problm na men, snadno eiteln koly a omezen dlky funkc na rozumn minimum  
    - ryzbyt na vce funkc, kde bude kad zodpovdn za jednu st problmu 
    - Obas znaeno jako `SRP` (Single Responsibility Principle)
3. Snaha o minimalizaci zanoen a sloitosti kdu.
4. Pouvn koment a dokumentace pro vysvtlen sloitch st kdu.

2. Rozdlen problm na men, snadno eiteln koly, rozdlen zodpovdnosti funkc


```python
# patn 

def vypocti_prumer_a_vypis(cisla):
    suma = sum(cisla)
    prumer = suma / len(cisla)
    print(prumer)


cisla = [1, 2, 3, 4, 5]
vypocti_prumer_a_vypis(cisla)
```

```python
# dobe

def vypocti_prumer(cisla):
    suma = sum(cisla)
    prumer = suma / len(cisla)
    return prumer


cisla = [1, 2, 3, 4, 5]

prumer = vypocti_prumer(cisla)
print(prumer)
```

```python
# patn

def vypocet_pi_newton(n_rada, n_odmocnina):
    a = 1 / 16
    s = a / 3
    for i in range(2, n_rada + 1):
        a = a * (2 * i - 3) / 8 / (i)
        s += a / (2 * i + 1)

    val_squared = 3
    x = val_squared
    for i in range(n_odmocnina):
        x = (x + val_squared / x) / 2

    odhad_pi = 12 * (-x / 8 + 1 / 2 - s)

    return odhad_pi

print(vypocet_pi_newton(10, 5))
```

```python
# dobe

def odmocnina(a, max_iter=10):
    x = a
    for i in range(max_iter):
        x = (x + a / x) / 2
    return x

def sum_rada_newton(n):
    a = 1 / 16
    s = a / 3
    for i in range(2, n + 1):
        a = a * (2 * i - 3) / 8 / (i)
        s += a / (2 * i + 1)
    return s

def vypocet_pi_newton(n_rada, n_odmocnina):
    s = sum_rada_newton(n_rada)
    sqrt_3 = odmocnina(3, n_odmocnina)
    odhad_pi = 12 * (-sqrt_3 / 8 + 1 / 2 - s)
    return odhad_pi

print(vypocet_pi_newton(10, 5))
```

3. Snaha o minimalizaci zanoen a sloitosti kdu.


```python
def slozita_kontrola(x, y):
    if x > 0:
        if y > 0:
            return x + y
        else:
            return x - y
    else:
        if y > 0:
            return y - x
        else:
            return -x - y
        
print(slozita_kontrola(1, 2))
print(slozita_kontrola(1, -2))
print(slozita_kontrola(-1, 2))
print(slozita_kontrola(-1, -2))
```

```python
def jednoducha_kontrola(x, y):
    if x > 0 and y > 0:
        return x + y
    elif x > 0:
        return x - y
    elif y > 0:
        return y - x
    else:
        return -x - y

print(jednoducha_kontrola(1, 2))
print(jednoducha_kontrola(1, -2))
print(jednoducha_kontrola(-1, 2))
print(jednoducha_kontrola(-1, -2))
```

4. Pouvn koment a dokumentace pro vysvtlen sloitch st kdu.

```python
# patn

def det(A):
    return A[0][0] * A[1][1] - A[0][1] * A[1][0]
```

```python
# dobe
from typing import List

def determinant(matice: List[List[float]]) -> float:
    """Spot determinant 2x2 matice.

    Args:
        matice (List[List[float]]): matice 2x2 v listu

    Returns:
        float: determinant matice
    """
    return matice[0][0] * matice[1][1] - matice[0][1] * matice[1][0]
```

## YAGNI (You Aren't Gonna Need It)

zamuje se na neimplementovn funkc a st kdu, kter nejsou aktuln poteba, a na odolvn pokuen pidvat do kdu funkcionalitu "pro jistotu" nebo "pro budouc poteby".

Neznamen to vak, e nemte mt promylen nvrh tak, aby budouc rozen bylo provediteln bez pepsn existujcho kdu.

Typicky to znamen:
- pedasn optimalizace
- pedasn a plin abstrakce
- generalizace kdu, pro funkcionality, kter nejsou aktuln poteba


# LoD (Law of Demeter) 
je nvrhov princip pouvan v objektov orientovanm programovn, kter se zamuje na snen zvislost mezi objekty a jejich spoluprci. Clem tohoto principu je omezit, jak moc jednotliv objekty vd o struktue a chovn ostatnch objekt, co zlepuje modularitu a sniuje zvislost mezi komponentami.

LoD stanovuje, e metoda objektu by mla komunikovat pouze s:

- Samotnm objektem.
- Parametry, kter pijm metoda.
- Jakmkoli objektem, kter metoda vytvo nebo instancuje.
- Jakmkoli objektem, kter je pmo soust objektu (jeho atributy).

V na praxi to hlavn znamen, e je teba dobe rozmyslet kam dan metoda pat. Sname se vyhnout "maitanance" tdm/funkcm, kter maj v sob functionality, kter by mly bt v jinch tdch/funkcch.

# 1. Práce s `venv` pro tento týden

Tento notebook je úvod ke všem dalším částem týdne 11. Cílem je mít jedno stabilní virtuální prostředí (`venv`) pro profilování, NumExpr, Numbu, Cython a související nástroje.


## 1.1 Proč používat `venv`

- oddělíme balíčky kurzu od systémového Pythonu,
- každý student může mít stejné verze nástrojů.


## 1.2 Vytvoření a aktivace prostředí

V terminálu v kořeni repozitáře:

```bash
python -m venv .venv
source .venv/bin/activate
python -m pip install --upgrade pip
python -m pip install -r .devcontainer/requirements-week11.txt
```

Opakování: když otevřete nový terminál, prostředí je potřeba znovu aktivovat (`source .venv/bin/activate`).


## 1.3 Jupyter kernel z `venv`

```bash
python -m ipykernel install --user --name vvp-venv --display-name "Python (vvp .venv)"
```

V prostředí Jupyter notebooků pak pro týden 11 vybírejte kernel `Python (vvp .venv)`.


````python
import sys
print(sys.executable)
````

## 1.4 Rychlé opakování práce s `venv`

1. Aktivuj: `source .venv/bin/activate`.
2. Ověř Python: `python -c "import sys; print(sys.executable)"`.
3. Spouštěj notebooky s kernelem `Python (vvp .venv)`.
4. Když něco chybí, doinstaluj to do `.venv`, ne do systému.
